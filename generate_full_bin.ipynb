```python
import random
#generate full data
def generate_full_bin():
  #width
  #randomly choose 2 integer in 1-5
  x_cut_point = []
  x_cut_point = random.sample(range(1,6),2)
  x_cut_point.sort()
  print("x:",x_cut_point)

  width_data = []
  for now in range(len(x_cut_point)):
    if now == 0:
      width_data.append(x_cut_point[now]*100)
    else:
      width_data.append(x_cut_point[now]*100 - x_cut_point[now-1]*100)
  width_data.append(600 - x_cut_point[len(x_cut_point)-1]*100)
  print("width_data:",width_data)

  #height
  #generate 3 set of y cutpoint total 9 cutpoint
  y_1 = random.randint(0,9)
  y_2 = random.randint(0,9-y_1)
  y_3 = 9 - y_1 - y_2
  print(y_1,y_2,y_3,"total:",y_1+y_2+y_3)
  #generate first collumn height data
  if y_1 > 0:
    y1_cut_point = []
    y1_cut_point = random.sample(range(1,10),y_1)
    y1_cut_point.sort()
    print("y1:",y1_cut_point)

    height_data1 = []
    for now in range(len(y1_cut_point)):
      if now == 0:
        height_data1.append(y1_cut_point[now]*100)
      else:
        height_data1.append(y1_cut_point[now]*100 - y1_cut_point[now-1]*100)
    height_data1.append(1000 - y1_cut_point[len(y1_cut_point)-1]*100)
    
  else:
    height_data1 = [1000]
  
  #generate second collumn height data
  if y_2 > 0:
    y2_cut_point = []
    y2_cut_point = random.sample(range(1,10),y_2)
    y2_cut_point.sort()
    print("y2:",y2_cut_point)

    height_data2 = []
    for now in range(len(y2_cut_point)):
      if now == 0:
        height_data2.append(y2_cut_point[now]*100)
      else:
        height_data2.append(y2_cut_point[now]*100 - y2_cut_point[now-1]*100)
    height_data2.append(1000 - y2_cut_point[len(y2_cut_point)-1]*100)
    
  else:
    height_data2 = [1000]

  #generate third collumn height data
  if y_3 > 0:
    y3_cut_point = []
    y3_cut_point = random.sample(range(1,10),y_3)
    y3_cut_point.sort()
    print("y3:",y3_cut_point)

    height_data3 = []
    for now in range(len(y3_cut_point)):
      if now == 0:
        height_data3.append(y3_cut_point[now]*100)
      else:
        height_data3.append(y3_cut_point[now]*100 - y3_cut_point[now-1]*100)
    height_data3.append(1000 - y3_cut_point[len(y3_cut_point)-1]*100)
    
  else:
    height_data3 = [1000]
  print("height_data1:",height_data1)
  print("height_data2:",height_data2)
  print("height_data3:",height_data3)

  #combine width and height data to generate all 12 slice data string
  slice_data_string = ""
  for now in range(len(height_data1)):
    slice_data_string = slice_data_string + str(width_data[0]) + " " + str(height_data1[now]) + "\n"
  for now in range(len(height_data2)):
    slice_data_string = slice_data_string + str(width_data[1]) + " " + str(height_data2[now]) + "\n"
  for now in range(len(height_data3)):
    slice_data_string = slice_data_string + str(width_data[2]) + " " + str(height_data3[now]) + "\n"

  #print("slice_data_string:")
  #print(slice_data_string)
  return slice_data_string
generated_string = generate_full_bin()
print("full bin:\n",generated_string)
```
